```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
URL ="https://github.com/ageron/handson-ml2/blob/master/datasets/housing/housing.csv?raw=true"
df = pd.read_csv(URL)
df.head()
```

```python
bins=[0,1.5, 3.0, 4.5, 6.0, np.inf]
labels=[1,2,3,4,5]
df['income_cat']=pd.cut(df['median_income'], bins=bins, labels=labels)
df.head()
```

```python
df['income_cat'].value_counts()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
sss=StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=45)

for train_index, test_index in sss.split(df, df['income_cat']):
  train_set=df.loc[train_index]
  test_set=df.loc[train_index]
```

OrdinalEncoder
Turli categorylardagi matnlarni son ko'rinishiga o'tkazadi





```python

from sklearn.preprocessing import OrdinalEncoder

# Bu modul categorylarni songa ajratib beradi.
#housing_cat=new_housing[['ocean_proximity']]

# Bizning dfdagi matnlarni songa o'tkazib beradi
ordinal_encoder=OrdinalEncoder()
```

```python
data={'A':[23,1,8,72,13,66,45,14],
      'B':['Red', 'Black','Black','Red','Green','Red','Black','Green']}
color_df=pd.DataFrame(data)
color_df
```

```python
new_color=ordinal_encoder.fit_transform(color_df[['B']]) # bu metodning ichiga df kiritilishi kerak Series emas
# shuning uchun color_df[['B']] quyidagicha df holatiga keltirib olamiz
```

```python
new_color
```

```python
color_df['C']=new_color
```

**Lekin bu usulning kamchiligi** - sonlar qancha bir biriga yaqin bo'lsa algoritm ularni shunchalik bir biriga yaqin deb o'ylaydi. Bizning misolda Black - 0, Green - 1, Red - 2 bo'ldi. Algoritm esa 0 soni 1 ga yaqin bo'lgani uchun Black ni Greenga yaqin orasida bog'liqlik bor deb o'ylaydi. Lekin haqiqatda esa bu xato. Bular shunchaki ranglar va ularning bir biri bilan hech qanday aloqasi yo'q. Bu metodning kamchiligi shunda.

**OneHotEncoder**


Bu ham categoriyalarni son ko'rinishiga olib o'tadi. Array qaytaradi.Lekin necha xil category bo'lsa o'shancha ustun qaytarib, mos keluvchi ustunga 1, qolganlariga 0 qo'yadi.

```python
color_df
```

```python
from sklearn.preprocessing import OneHotEncoder

one_encoder=OneHotEncoder()
# Bu metodga ham 2d array kiritilishi lozim

one_hot=one_encoder.fit_transform(color_df[['B']])

# Keyin esa bu qaytgan javobni toarray() orqali arrayga o'tkazib olamiz va yana qayta dfga o'tkazamiz
```

```python
abc=one_hot.toarray()
abc
```

```python
color_df[['Black','Green','Red']]=abc  # endi bu arrayni dfning davomiga qo'shib qo'yamiz
```
