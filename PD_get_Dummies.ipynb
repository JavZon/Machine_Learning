```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

**pd.get_dummies()** - Bu usul bilan ham biz matnli valuelarni sonli ustunlarga almashtirishimiz mumkin bo'ladi.

```python
data={'Animal':['cat','dog','sheep','cow','cat']}
df=pd.DataFrame(data)
```

```python
pd.get_dummies(df)   # Hech nima yozmasak dtype='bool' bo'lib bizga True/False qaytaradi
```

```python
pd.get_dummies(df, dtype='int')  # dtype='int' qilsak, bizga 1,0 qiymatlardan iborat columnlarni qaytaradi
```

```python
from sklearn.preprocessing import OneHotEncoder
one_enc=OneHotEncoder()
A=one_enc.fit_transform(df).toarray()
```
