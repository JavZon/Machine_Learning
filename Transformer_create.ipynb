```python
class Owner:
  def __init__(self, name, age, passport):
    self.name=name
    self.age=age
    self.passport=passport

  def owner_info(self):
    return f"Info about owner: Name - {self.name}, Age - {self.age}, Passport - {self.passport}"
```

```python
class Car:
  def __init__(self, brand, color,year,price,owner):
        self.brand = brand
        self.color = color
        self.year=year
        self.price=price
        self.owner=owner

  def start_engine(self):
        return f"The {self.color} {self.brand} car has started."
  def info(self):
        return f"Info about car : Brand - {self.brand}, color - {self.color}, year - {self.year}, price - {self.year}."
  def get_owner(self):
      return self.owner.owner_info()
```

**BaseEstimator and TransformerMixin**

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
data={'A':[78,34,21,44,90,3,1],
      'B':[312,68,21,110,45,9,0],
      'C':[56,16,20,61,9,12,83]}

df=pd.DataFrame(data)
df
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class CustomTransformer(BaseEstimator, TransformerMixin):
  def __init__(self):
    pass
  def fit(self, X, y=None):
    return self
  def transform(self, X):
    X['B/A_Transformed_Col']=(X['B']/X['A'])*2
    return X
```

```python
cus_obj=CustomTransformer()
A=cus_obj.fit_transform(df)
```
